```python
# import packages
from __future__ import print_function, division
from IPython.display import display, clear_output, Image

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import seaborn as sns
import pprint

from sklearn import linear_model
from sklearn.model_selection import train_test_split
from statsmodels.api import OLS # Ordinary Least Squares 普通最小二乘模型
from sklearn import ensemble
from sklearn.model_selection import cross_val_predict
```

```python
fname = "E:/JI/2021SU/VG441/Midterm/demands.csv"
df = pd.read_csv(fname)
# df.head()
```

```python
time = df.iloc[:,0:1]
demands = df.iloc[:,1:2]

plt.title("Figure 1")
plt.xlabel("time")
plt.ylabel("demands")
plt.scatter(time,demands)
plt.show()
```

```python
model = linear_model.LinearRegression()
model.fit(time,demands)

demands2 = model.predict(time)

plt.title("Figure 2")
plt.xlabel("time")
plt.ylabel("demands")

plt.scatter(time,demands)
plt.plot(time,demands2)
plt.show()
```

```python
params = {'n_estimators': 1, 'max_depth': 1, 'learning_rate': 1, 'loss': 'ls'}
model = ensemble.GradientBoostingRegressor(**params)
model.fit(time, demands)

demands2 = model.predict(time)

plt.scatter(time,demands)
plt.plot(time, demands2,c='r')

plt.title("Figure 3")
plt.xlabel("time")
plt.ylabel("demands")

plt.show()
```
