```python
def get_deviance(array):
    avg = sum(array) / len(array)
    deviance = 0
    for num in array:
        deviance += (num - avg)**2
    return deviance
def get_SS(array,lam):
    SS = sum(array)**2/(len(array)+lam)
    return SS
```
