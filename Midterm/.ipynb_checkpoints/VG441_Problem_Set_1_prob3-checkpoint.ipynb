```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pprint
import xgboost as xgb
import scipy as sp

from sklearn import linear_model
from sklearn import ensemble
from sklearn.metrics import mean_squared_error, r2_score
from statsmodels.api import OLS # Ordinary Least Squares 普通最小二乘模型
from sklearn.model_selection import train_test_split, cross_val_predict
```

```python
fname = "E:/JI/2021SU/VG441/Problem_Sets/CALI.csv"
origin_df = pd.read_csv(fname)
# origin_df.head()
```

```python
"""get df_x and df_y"""
df_x = origin_df.iloc[:, 0:8]
df_y = origin_df.iloc[:, 8]
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size = 0.2, random_state = 4)
```

# Task 1: Linear Regression Model

```python
model_LR = OLS(endog = df_y, exog = df_x).fit() #
model_LR.summary()
```

```python
model = linear_model.LinearRegression()
model.fit(x_train,y_train)
results = model.predict(x_test) # len(results) == 4128
```

```python
"""check the first 5 items of y_test and predict results"""
five_item = np.c_[y_test.values, results][0:5,:]
# print(five_item)
```

```python
model_score = model.score(x_train,y_train)
print('R2 sq for training data: ', model_score)

# The mean squared error
print("Mean squared error: %.2f"% mean_squared_error(y_test, results))

# Explained variance score: 1 is perfect prediction
print('Test Variance score: %.2f' % r2_score(y_test, results))
```

```python
pred_val = model_LR.fittedvalues.copy()
residual = pred_val - df_y.values.flatten()

fig, ax = plt.subplots(figsize=(6,2.5))
ax.set_xlabel('residual')
ax.set_ylabel('pred_val')
ax.scatter(residual, pred_val)
```

```python
sns.distplot(residual);
```

```python
fig, ax = plt.subplots(figsize=(6,2.5))
_, (__, ___, r) = sp.stats.probplot(residual, plot=ax, fit=True)
```

```python
# Examining simple linear regression...
for i in range(1):
    X = df_x.iloc[:,i].values.reshape(-1, 1)
    Y = df_y.values.reshape(-1,1)

    model3 = linear_model.LinearRegression()
    model3.fit(X,Y)
    Y_pred = model3.predict(X)

    plt.scatter(X, Y)
    plt.plot(X, Y_pred, color='red')
    plt.show()
```

## Interpretation

- OLS Regression Results：

    - `coef` of `Population` is `7.792e-06`, which is very small. This may suggest population's impact on housing prise is negligible.
    
- For training data:  
    - R2 sq  = 0.8043405623940919, which is also relatively close to 1.
    - Mean squared error = 0.53

    - These all implies that the linear model is relatively appropriate for this dataset.

# Task 2: GradBoost

```python
# Gradient boosting in action...
# n_estimators：m in the slides
# max_depth: four layers
# min_samples_split: 
params = {'n_estimators': 1000, 'max_depth': 4, 'min_samples_split': 2, 'learning_rate': 0.01, 'loss': 'ls'}
model = ensemble.GradientBoostingRegressor(**params)
model.fit(x_train, y_train)
```

```python
model_score = model.score(x_train,y_train)
print('R2 sq: ',model_score)

y_predicted = model.predict(x_test)
# The mean squared error
print("Mean squared error: %.2f"% mean_squared_error(y_test, y_predicted))
# Explained variance score: 1 is perfect prediction
print('Test Variance score: %.2f' % r2_score(y_test, y_predicted))
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, y_predicted, edgecolors=(0, 0, 0))
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title("Ground Truth vs Predicted")
plt.show()
```

## Interpretation
- R2 sq = 0.8391271074833364, which is relatively close to 1
- Mean squared error = 0.25, which is relatively small
- Test Variance score = 0.81, which is relatively close to 1
- These all implies that the linear model is relatively appropriate for this dataset.

# Task 3: XGBoost

```python
# XGBoost in action...
params = {'n_estimators': 500, "objective":"reg:linear",'colsample_bytree': 0.5,'learning_rate': 0.01,'max_depth': 4, 'alpha': 1}
xg_reg = xgb.XGBRegressor(**params)

xg_reg.fit(x_train,y_train)
```

```python
y_predicted = xg_reg.predict(x_test)
# The mean squared error
print("Mean squared error: %.2f"% mean_squared_error(y_test, y_predicted))
# Explained variance score: 1 is perfect prediction
print('Test Variance score: %.2f' % r2_score(y_test, y_predicted))
```

```python
# Gradient boosting in action...
params = {'n_estimators': 500, 'max_depth': 4, 'min_samples_split': 2, 'learning_rate': 0.01, 'loss': 'ls'}
model = ensemble.GradientBoostingRegressor(**params)
model.fit(x_train, y_train)
y_predicted_2 = model.predict(x_test)
# The mean squared error
print("Mean squared error: %.2f"% mean_squared_error(y_test, y_predicted_2))
# Explained variance score: 1 is perfect prediction
print('Test Variance score: %.2f' % r2_score(y_test, y_predicted_2))
```

```python
#XGBoost Visualizations
params = {"objective":"reg:linear",'colsample_bytree': 0.5,'learning_rate': 0.01, 'max_depth': 4, 'alpha': 1}
data_dmatrix = xgb.DMatrix(data=df_x,label=df_y)
xg_reg = xgb.train(params=params, dtrain=data_dmatrix, num_boost_round=10)
xgb.plot_tree(xg_reg,num_trees=9)
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.show()
```

```python
xgb.plot_importance(xg_reg)
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.show()
```

## Interpretation
- Mean squared error = 0.28, which is relatively small
- Test Variance score = 0.78, which is relatively close to 1
- These all implies that the linear model is relatively appropriate for this dataset.
